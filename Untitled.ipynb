```python
with open("r_300_2") as f:
    lines = f.readlines()
para2 = []
result2 = []
for i in range(0,len(lines),2):
    para2.append(np.float32(re.findall(r'\d+\.?\d*e?[-+]?\d+', lines[i+1])))
    result2.append(np.float32(re.findall(r'\d+\.?\d*e?[-+]?\d+', lines[i])))
for i in para2:
    hidden_layer_1.append(i[0])
    hidden_layer_2.append(i[1])
    r_para.append(i[4])
    if len(i) == 8:
        momentum.append(i[5])
        learning_rate.append(i[7])
        batch_size.append(i[6])
    else:
        momentum.append(0)
        learning_rate.append(i[6])
        batch_size.append(i[5])
for i in result2:
    acc_avg.append(i[0])
    acc_std.append(i[1])
    loss_avg.append(i[2])
    loss_std.append(i[3])
```

```python
with open("r_300_3") as f:
    lines = f.readlines()
para3 = []
result3 = []
for i in range(0,len(lines),2):
    para3.append(np.float32(re.findall(r'\d+\.?\d*e?[-+]?\d+', lines[i+1])))
    result3.append(np.float32(re.findall(r'\d+\.?\d*e?[-+]?\d+', lines[i])))

for i in para3:
    hidden_layer_1.append(i[0])
    hidden_layer_2.append(i[1])
    learning_rate.append(i[2])
    if len(i) == 8:
        momentum.append(i[3])
        r_para.append(i[4])        
        batch_size.append(i[7])
    else:
        momentum.append(0)
        r_para.append(i[3])   
        batch_size.append(i[6])

for i in result3:
    acc_avg.append(i[0])
    acc_std.append(i[1])
    loss_avg.append(i[2])
    loss_std.append(i[3])
```

```python
import pandas as pd
import plotly.express as px
df = []
df = pd.DataFrame(data = hidden_layer_1,columns=['hidden_layer_1'])
df['r_para'] = r_para
df['learning_rate'] = learning_rate
df['momentum'] = momentum
df['hidden_layer_2'] = hidden_layer_2
df['batch_size'] = batch_size

df['acc_avg'] = acc_avg
df['acc_std'] = acc_std
df['loss_avg'] = loss_avg
df['loss_std'] = loss_std

fig = px.parallel_coordinates(df,
                              dimensions=['hidden_layer_1', 'r_para',
                                          'learning_rate', 'momentum','hidden_layer_2','batch_size','acc_avg','acc_std',
                                        'loss_avg', 'loss_std'],
                              color_continuous_scale=px.colors.diverging.Tealrose,
                              color_continuous_midpoint=2)
fig.show()
```
